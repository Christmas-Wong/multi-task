```python
import pandas as pd
import os
from pandas import DataFrame
from typing import Tuple


DIR_SOURCE = "/Users/wangfei/Documents/data/open_source_data/classification/asap-master/asap"
DIR_OUTPUT = "/Users/wangfei/Documents/project/server_multitask/data/origin"
# train.csv,eval.csv,test.csv
```

```python
df_train = pd.read_csv(os.path.join(DIR_SOURCE, "train.csv"), encoding="utf-8")
df_test = pd.read_csv(os.path.join(DIR_SOURCE, "test.csv"), encoding="utf-8")
df_train
```

```python
print(df_train["Location#Transportation"].value_counts())
print(df_train["Service#Queue"].value_counts())
```

```python
def data_builder(df: DataFrame) -> Tuple[DataFrame, DataFrame]:
    """Build Data
    1. Choose review for text
    2. Choose Location#Transportation & Service#Queue for two different task
    
    Args:
        df (DataFrame): Source Data
    Returns:
        df_location (DataFrame): Data of location label
        df_service (DataFrame): Data of service label
    """
    dict_location = {
        "text": list(df["review"]),
        "label": list(df["Location#Transportation"])
    }
    dict_service = {
        "text": list(df["review"]),
        "label": list(df["Service#Queue"])
    }
    df_location = DataFrame(dict_location)
    df_service = DataFrame(dict_service)
    print("DataFrame has [{}]rows & [{}]columns".format(df_service.shape[0], df_service.shape[1]))
    return df_location, df_service
```

```python
df_location_train, df_service_train = data_builder(df_train)
df_location_test, df_service_test = data_builder(df_test)
df_service_test
```

```python
df_location_train.to_csv(os.path.join(DIR_OUTPUT, "location_train.csv"), encoding="utf-8", index=False)
df_location_test.to_csv(os.path.join(DIR_OUTPUT, "location_test.csv"), encoding="utf-8", index=False)
df_service_train.to_csv(os.path.join(DIR_OUTPUT, "service_train.csv"), encoding="utf-8", index=False)
df_service_test.to_csv(os.path.join(DIR_OUTPUT, "service_test.csv"), encoding="utf-8", index=False)
```

```python
import pickle
from sklearn.preprocessing import LabelEncoder

def generate_label_encoder(df: DataFrame, label: str, save_path) -> LabelEncoder:
    """Generate Label Encoder & save to file

    Args:
        df (DataFrame): Train DataFrame
        label (str): Column Name of Label
        save_path ([type]): File Path to Save

    Returns:
        LabelEncoder: Label Encoder
    """
    if os.path.exists(save_path):
        return get_label_encoder(save_path)
    
    le = LabelEncoder()
    le.fit(df[label].tolist())
    output = open(save_path, 'wb')
    pickle.dump(le, output)
    return le
```

```python
generate_label_encoder(df_location_train, "label", os.path.join(DIR_OUTPUT, "0_encoder.pkl"))
generate_label_encoder(df_service_train, "label", os.path.join(DIR_OUTPUT, "1_encoder.pkl"))
```
